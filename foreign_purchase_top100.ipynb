```python
from bs4 import BeautifulSoup
import requests

def get_html(url):
    _html = ""
    resp = requests.get(url)
    if resp.status_code == 200:
        _html = resp.text
    return _html


URL = "http://finance.naver.com/sise/sise_deal_rank_iframe.nhn?sosok=01&investor_gubun=9000&type=buy&page=1"
html = get_html(URL)
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
Code_text = soup('p',{'class':'tit'},)
Code_text
```

```python
for t in Code_text:
    codeText=str(t)
    print(codeText.split('code=')[1])
```

```python
for t in Code_text:
    codeText=str(t)
    print(codeText.split('code=')[1].split('"')[0],
         codeText.split('code=')[1].split('title="')[1].split('">')[0])
```

```python
code_list = []
name_list = []
for i in range(50):
    URL = "http://finance.naver.com/sise/sise_deal_rank_iframe.nhn?sosok=01&investor_gubun=9000&type=buy&page="+str(i+1)
    html = get_html(URL)
    soup = BeautifulSoup(html, 'html.parser')
    Code_text = soup('p',{'class':'tit'},)
    for t in Code_text:
        codeText=str(t)
        code_list.append(codeText.split('code=')[1].split('"')[0])
        name_list.append(codeText.split('code=')[1].split('title="')[1].split('">')[0])
```

```python
import pandas as pd
import numpy as np
```

```python
data = {'code': code_list,
       'name': name_list}
df = pd.DataFrame(data)
df.name.value_counts()
```
